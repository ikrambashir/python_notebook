```python
import pandas as pd
data = pd.read_csv('potato.CSV')
data.head()
```

```python
# Adding a Y-Axis Label to the
# Secondary Y-Axis in Matplotlib
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt

# # creating data for plot
# # data arrangement between 0 and 50 with the difference of 2
# # x-axis values
# x = np.arange(0, 50, 2)
# #y-axis values
# y1 = x**2
# # secondary y-axis values
# y2 = x**3
import seaborn as sns
sns.set_theme(style="dark")
# plotting figures by creating aexs object
# using subplots() function
fig, ax = plt.subplots(figsize = (10, 5))
plt.title('Effect of Temp on Potato 1960-2020 in Pakistan', weight= 'bold')

# using the twinx() for creating
# another axes object for secondary y-Axis
ax2 = ax.twinx()
# creating a bar plot
ax.plot("Year", "Yield", color = 'g', data=data)
ax2.plot("Year", "°C", color = 'b', data=data)

# giving labels to the axises
ax.set_xlabel('Year', color = 'r')
ax.set_ylabel('Yield (matric ton)', color = 'g')

# secondary y-axis label
ax2.set_ylabel('Temperature (°C)', color = 'orange')

# defining display layout
plt.tight_layout()

# show plot
plt.show()
```
